```python
import os
import shutil
```

```python
if os.path.exists('WFLW_anno.txt'):
    os.remove('WFLW_anno.txt')
anno_file = open('WFLW_anno.txt', 'a')
```

```python
train_landmark_path = "/Users/yuhua.cheng/Downloads/WFLW_annotations/list_98pt_rect_attr_train_test/list_98pt_rect_attr_train.txt"
```

```python
with open(train_landmark_path, 'r') as f:
    landmarks = f.readlines()
    for _ in landmarks:
        label = _.strip().split(' ')
        landmark = label[0:196]
        bbox = label[196:200]
        name = label[-1]
        print(bbox)
        print(name)
        anno_file.write(name+' ')
        anno_file.write(' '.join(bbox)+' ')
        anno_file.write(' '.join(landmark)+'\n')
```
